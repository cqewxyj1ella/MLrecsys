## 去除重复行

```python
import pandas as pd

path = '../data/train.csv'
data = pd.read_csv(path,encoding='utf-8')
pre_user_id = 0
pre_product = 0
pre_event = '0'
del_list = []

# 去除重复行
for index , row in data.iterrows():
    if row['user_id'] == pre_user_id:
        if pre_product == row['product_id'] and pre_event == row['event_type']:
            del_list.append(index)
            print(index)
    if row['user_id'] != pre_user_id:
        pre_user_id = row['user_id']
    pre_event = row['event_type']
    pre_product = row['product_id']

    
data.drop(del_list,axis=0,inplace=True)
data = data.reset_index(drop=True)
data.to_csv("../data/cleaned_train.csv",index=False)
```
